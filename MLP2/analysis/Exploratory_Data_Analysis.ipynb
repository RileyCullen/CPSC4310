# Dispersion

```python
name = ["BMI", "AgeCategory", "PhysicalHealth", "GenHealth", "SleepTime", "MentalHealth"]
colors = ["b", "g", "r", "c", "m", "y"]
numbins = [10, 13, 10, 5, 10, 10]
fig, ax = plt.subplots(1, 6, figsize=(22, 8), layout="tight")
for i in range(len(name)):
    ax[i].hist(df_2020[name[i]], label=name[i], color=colors[i], bins=numbins[i])
    ax[i].set_title(name[i])
    ax[i].set_ylabel("Frequency")
plt.show()
```

 We can see from these histograms BMI, SleepTime, and GenHealth are all Unimodal and have relatively normal distributions with BMI and SleepTime having slight right skews. AgeCategory looks to be trimodal with peaks at both ends of the attribute and a peak around 8.0. Physical health is unimodal witha a large right skew. Most of the distribution lies between 0.0 and 0.2. This is very similar to the MentalHealth distribution.

```python
#Get binary value columns drop the rest
df_bin = df_2020.drop(columns=["BMI", "AgeCategory", "PhysicalHealth", "GenHealth", "SleepTime", "MentalHealth"])
df_bin.columns
```

```python
#Plot Count of binary attributes
fig, ax = plt.subplots(4, 5, figsize=(16, 12), layout="tight")
index = 0
row = 0
for (columnName, columnData) in df_bin.iteritems():
    ax[row][index].hist(df_bin[columnName], bins=3)
    ax[row][index].set_title(columnName)
    ax[row][index].set_ylabel("Frequency")
    index = index + 1
    if(index == 5):
        index = 0
        row = row + 1
```

```python
name = ["trestbps", "chol", "thalch", "oldpeak", "ca", "age"]
colors = ["b", "g", "r", "c", "m", "y"]
numbins = [10, 10, 10, 10, 5, 10]
fig, ax = plt.subplots(1, 6, figsize=(20, 8), layout="tight")
for i in range(len(name)):
    ax[i].hist(df_uci[name[i]], label=name[i], color=colors[i], bins=numbins[i])
    ax[i].set_title(name[i])
    ax[i].set_ylabel("Frequency")
plt.legend()
plt.show()
```

trestbps, chol, thalc, and age all have relatively normal distributions and all are unimodal except for chol which is bimodal. trestbps also has a slight left skew while chol has a slight right skew. oldpeak has a large frequency between 0.2 and 0.3. The ca histogram is bimodal and right skewed with most of the distribution between 0.2 and 0.4
